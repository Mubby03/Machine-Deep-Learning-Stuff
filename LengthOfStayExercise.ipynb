```python
# Load the dataset from Google Drive into a pandas DataFrame
df = pd.read_csv("/content/drive/MyDrive/Lengthofstay.csv")
# Display the first few rows of the DataFrame to get an overview of the data
df.head()
```

```python
# Display the names of all columns in the DataFrame
df.columns
```

```python
# Generate descriptive statistics of the DataFrame
# .T (transpose) is used to make the output more readable, showing statistics per column
df.describe().T
```

```python
# Display the number of rows and columns in the DataFrame
df.shape
```

```python
# Check for missing values in each column and sum them up
df.isnull().sum()
```

```python
# Initialize the StandardScaler to normalize the data
scaler = StandardScaler()
# Fit the scaler to the DataFrame and transform the data
Clus_dataSet = scaler.fit_transform(df)
# Convert the scaled data back into a pandas DataFrame for easier manipulation
scaled_features_clus = pd.DataFrame(Clus_dataSet )
```

##Kmeans Clustering

> Add blockquote

```python
# Define the number of clusters for KMeans
kclusters = 3
# Initialize KMeans model with the specified number of clusters and a random state for reproducibility
kmeans = KMeans(n_clusters=kclusters, random_state=42)
# Fit the KMeans model to the scaled dataset
kmeans.fit(Clus_dataSet)
```

```python
# Display the first 100 cluster labels assigned by KMeans
kmeans.labels_[0:100]
```

```python
# Count the number of data points assigned to each cluster
pd.Series(kmeans.labels_).value_counts()
```

```python
# Insert the KMeans cluster labels as a new column at the beginning of the original DataFrame
df.insert(0, 'kmeans Cluster Labels', kmeans.labels_)
# Display the first 100 rows of the modified DataFrame to see the new cluster labels
df.head(100)
```

```python
# Calculate and print the Silhouette Score for the KMeans clustering
# The Silhouette Score measures how similar an object is to its own cluster compared to other clusters
print('Silhouette Score:', '%.2f'%silhouette_score(Clus_dataSet, kmeans.labels_))
```

```python
# Add the KMeans cluster labels to the DataFrame containing scaled features
scaled_features_clus['cluster'] = kmeans.labels_
```

```python
# Create separate DataFrames for each cluster based on KMeans labels
cluster1=scaled_features_clus.query("cluster == 0")
cluster2=scaled_features_clus.query("cluster == 1")
cluster3=scaled_features_clus.query("cluster == 2")
```

```python
# Display the first few rows of cluster 1
cluster1.head()
```

```python
# Generate descriptive statistics for cluster 1
cluster1.describe()
```

```python
# Display the first few rows of cluster 2
cluster2.head()
```

```python
# Assign the coordinates of each cluster's center to individual variables
cluster1_mean = kmeans.cluster_centers_[0]
cluster2_mean = kmeans.cluster_centers_[1]
cluster3_mean = kmeans.cluster_centers_[2]
```

## kmedoids Clustering

## Hierarchical Clustering

```python
# Initialize AgglomerativeClustering model with 3 clusters, 'euclidean' metric, and 'complete' linkage
# Linkage options include 'ward', 'complete', 'average', 'single'
hcluster = AgglomerativeClustering(n_clusters=3, metric='euclidean',linkage='complete')
# Fit the Hierarchical clustering model to the scaled dataset and predict cluster labels
hcluster.fit_predict(Clus_dataSet)
```

```python
# Display the first 100 cluster labels assigned by Hierarchical Clustering
hcluster.labels_[0:100]
```

```python
# Count the number of data points assigned to each cluster by Hierarchical Clustering
pd.Series(hcluster.labels_).value_counts()
```

```python
# Insert the Hierarchical clustering labels as a new column at the beginning of the original DataFrame
df.insert(0, 'hcluster Cluster Labels', hcluster.labels_)
# Display the first few rows of the modified DataFrame to see the new cluster labels
df.head()
```

```python
# Calculate and print the Silhouette Score for the Hierarchical clustering
print('Silhouette Score:', '%.2f'%silhouette_score(Clus_dataSet, hcluster.labels_))
```

```python
# Add the Hierarchical clustering labels to the DataFrame containing scaled features
scaled_features_clus['cluster'] = hcluster.labels_
```

```python
# Create separate DataFrames for each cluster based on Hierarchical clustering labels
cluster1=scaled_features_clus.query("cluster == 0")
cluster2=scaled_features_clus.query("cluster == 1")
cluster3=scaled_features_clus.query("cluster == 2")
```
